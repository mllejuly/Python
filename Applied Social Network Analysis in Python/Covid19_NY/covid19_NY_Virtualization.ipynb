# NY Choropleth Map

```python
# Import Package
import pandas as pd
import json
import plotly.express as px
import numpy as np
```

```python
# Load Geography Json
from urllib.request import urlopen
import json

# with urlopen('https://raw.githubusercontent.com/plotly/datasets/master/geojson-counties-fips.json') as response:
#     us_counties = json.load(response)

# open JSON file
f = open('us_counties_geojson.json',) 

# returns JSON object as a dictionary 
us_counties_geojson = json.load(f) 
```

```python
# Read Data
df = pd.read_csv("covid19_NY_CountiesData.csv", dtype={"fips": str})
# df = df[df['date'] == '4/30/2020']
df
```

```python
# Visualization
fig = px.choropleth_mapbox(df,
                           geojson = us_counties_geojson,
                           color = 'confirmed',
                           range_color=(0, 30000),
                           # match location in csv with featureidkey in geoJson
                           locations = 'fips',
                           featureidkey = "properties.GEO_ID",
                           mapbox_style="carto-darkmatter",
                           color_continuous_scale='Reds',
                           center={"lat": 42.85, "lon": -75.5},
                           zoom = 5.3,
                           labels={'fips':'County',
                                   'confirmed':'Confirmed Cases'},
                           animation_frame="date"
)

fig.update_layout(
    # figure title
    title_text = 'New York State Covid19 Confirmed Cases on April 30',
    # figure title position
    title_x = 0.5,
    
    geo = dict(
        showframe = False,
        showcoastlines = False,
        projection_type = 'equirectangular'
    )
)

fig.show()
```
