```python
# Import Package
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
# Load Data
df = pd.read_csv("covid19_NY_daily.csv", dtype={"fips": str})
df
```

```python
date = df['date']
cases = df['cases']
deaths = df['deaths']
tweets = df['tweets']
```

```python
y1 = cases
y2 = tweets

fig = plt.figure(facecolor='white')
fig.set_size_inches(12, 8)

ax = plt.gca()
ax.grid(which='major', axis='x', linestyle='')

# Confirmed - line
ax1 = fig.add_subplot(111)
ax1.plot(date, y1, '#E55D3F', marker='', linewidth=3)
# ax1.plot(x, y1, '#E55D3F', marker='', linewidth=3)
ax1.set_ylabel('Confirmed Cases',fontsize='15', color='#E55D3F')
ax1.set_title("Confirmed Cases vs. Daily Tweets", fontsize='20')

# Tweets - rec
ax2 = ax1.twinx()
ax2.bar(date, y2, color='#00acee', alpha=.7)
ax2.set_ylabel('Tweets Number',fontsize='15', color='#00acee')


# y stick
plt.yticks(fontsize=10)

# x stick
index=[0,6,13,20,27]
x_index=['4/1','4/7','4/14','4/21','4/28']
plt.xticks(index, x_index, fontsize=10)
# x label
# plt.xlabel('Date')


plt.grid(False)
plt.show()
```
